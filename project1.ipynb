```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json 
import lxml
import re
url = 'https://www.spaceweatherlive.com/en/solar-activity/top-50-solar-flares'
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
response = requests.get(url, headers=headers)
content = response.content
```

```python
a = BeautifulSoup(content)
a.prettify()
solarTable = a.find('table')
table = pd.read_html(solarTable.prettify())
table=table[0]
table.columns = ['rank','x_class', 'date', 'region' ,'start_time', 'max_time', 'end_time', 'movie']
table
```
