```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json 
import lxml
url = 'https://www.spaceweatherlive.com/en/solar-activity/top-50-solar-flares' # sets URL for easy use
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
# sets headers to mock an actual user^
response = requests.get(url, headers=headers) # sends get request to spaceweather and collects response
content = response.content # sets the content of the response
```

```python
a = BeautifulSoup(content) # create a beautifulsoup object for the content 
a.prettify() # prettify it 
solarTable = a.find('table') # finds the table on the website 
dfSolar = pd.read_html(solarTable.prettify()) # converts html table into pandas df
dfSolar=dfSolar[0] 
dfSolar.columns = ['rank','x_class', 'date', 'region' ,'start_time', 'max_time', 'end_time', 'movie'] # sets columns
dfSolar
```

### Step 2 - Tidy Flare Data

```python
del dfSolar['movie'] # cuts 'movie' column
```

```python
import datetime
iterr = dfSolar.iterrows() # sets up iterator for table
for i,row in iterr: # loops through rows
    
    temp = row['start_time'] + ' '+ row['date'] # create one string with both start time and date
    startDatetime = pd.to_datetime(temp) # convert string into a datetime object 
    dfSolar.loc[i,'start_time'] = startDatetime # uses current row index to replace time with datetime object 
    
    temp = row['max_time'] + ' '+ row['date'] # repeats replacement for all types of times
    maxDatetime = pd.to_datetime(temp)
    dfSolar.loc[i,'max_time'] = maxDatetime
    
    temp = row['end_time'] + ' '+ row['date']
    endDatetime = pd.to_datetime(temp)
    dfSolar.loc[i,'end_time'] = endDatetime

dfSolar = dfSolar.drop(columns=['date'], axis=1) # drops date column since no longer needed
dfSolar = dfSolar.rename(columns={"start_time": "start_datetime", "max_time": "max_datetime", "end_time": "end_datetime"}, errors="raise")
# renames columns ^
dfSolar = dfSolar[['rank','x_class','start_datetime','max_datetime','end_datetime','region']] # rearrange columns
dfSolar
```

### Step 3 - Scrape Nasa Data

```python
import re
import numpy as np
url2 = 'https://cdaw.gsfc.nasa.gov/CME_list/radio/waves_type2.html'
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
response2 = requests.get(url2, headers=headers)
content2 = response2.content
```

```python
b = BeautifulSoup(content2, "html.parser")
s = str(b.pre).split('\n')[12:-2]
data = {"Start Date": [],
        "Start Time": [],
        "End Date": [],
        "End Time": [],
        "Max Frequency": [],
        "Min Frequency": [],
        "Flare Location": [],
        "NOAA": [],
        "Imp": [],
        "CME Date": [],
        "CME Time": [],
        "CPA": [],
        "Width": [],
        "Speed": []}
        
dfNasa = pd.DataFrame(data)
```

```python
for i in s:
    row = i.split(' ')
    a = [x for x in row if x]
    b = [y for y in a if y!='<a']
    temp = re.sub(r'href[^<]+>', '', b[4])
    noLinkMax = re.sub(r'</a>','', temp)
    noLinkMax = re.sub(r'</a>','', temp)
    if noLinkMax != '????':
        noLinkMax = int(noLinkMax)
    else:
        noLinkMax = np.nan
        
    
    temp = re.sub(r'href[^<]+>', '', b[5])
    noLinkMin = re.sub(r'</a>','', temp)
    if noLinkMin != '????':
        noLinkMin = int(noLinkMin)
    else:
        noLinkMin = np.nan
    
    tempCME = re.sub(r'href[^<]+>', '', b[9])
    noLinkCME = re.sub(r'</a>','', tempCME)
    
    tempSpeed = re.sub(r'href[^<]+>', '', b[13])
    noLinkSpeed = re.sub(r'</a>','', tempSpeed)
    
    imp = b[8]
    if imp[-1] == '.':
        imp = imp[:-1] + '+'
        
    year = b[0][0:4]
    enddate = year + '/'+ b[2]
    
    cmeDate = year + '/' + noLinkCME
    
    insertRow = pd.Series({"Start Date": b[0],
                       "Start Time": b[1],
                       "End Date": enddate,
                       "End Time": b[3],
                       "Max Frequency": noLinkMax,
                       "Min Frequency": noLinkMin,
                       "Flare Location": b[6],
                       "NOAA": b[7],
                       "Imp": imp,
                       "CME Date": cmeDate,
                       "CME Time": b[10],
                       "CPA": b[11],
                       "Width": b[12],
                       "Speed": noLinkSpeed
                      })
    dfNasa = dfNasa.append(insertRow, ignore_index=True)
dfNasa
```

### Step 4 - Tidy Nasa Data

```python
dfNasa["Flare Location"] = dfNasa["Flare Location"].replace("Back","BACK")
dfNasa["Flare Location"] = dfNasa["Flare Location"].replace("Back?","BACK")

dfNasa["NOAA"] = dfNasa["NOAA"].replace("-----",np.NaN)
dfNasa["NOAA"] = dfNasa["NOAA"].replace("DSF","FILA")
dfNasa["Imp"] = dfNasa["Imp"].replace("----",np.NaN)

dfNasa['CME Time'] = dfNasa['CME Time'].replace('--:--','00:00')

dfNasa["CME Date"] = dfNasa["CME Date"].replace("----",np.NaN)
dfNasa["CME Date"] = dfNasa["CME Date"].replace("----",np.NaN)
dfNasa["CPA"] = dfNasa["CPA"].replace("----",np.NaN)
dfNasa["Is Flare"] = [True if row == "Halo" else False for row in dfNasa['CPA']]
dfNasa["CPA"] = dfNasa["CPA"].replace("Halo",np.NaN)
dfNasa["Width"] = dfNasa["Width"].replace("----",np.NaN)
dfNasa["Speed"] = dfNasa["Speed"].replace("----",np.NaN)
dfNasa["Width Lower Bound"] = [True if "&gt;" in str(row) else False for row in dfNasa['Width']]
dfNasa["Width"] = [int(str(row)[4:]) if "&gt;" in str(row) else row for row in dfNasa['Width']]
dfNasa["End Time"] = dfNasa["End Time"].replace("24:00","23:59")
dfNasa["Start DateTime"] = pd.to_datetime(dfNasa["Start Date"] + " " + dfNasa["Start Time"])
dfNasa["End DateTime"] = pd.to_datetime(dfNasa["End Date"] + " " + dfNasa["End Time"])
dfNasa["CME DateTime"] = pd.to_datetime(dfNasa["CME Date"] + " " + dfNasa["CME Time"])

# Reorder table
"""
dfNasa = dfNasa[['Start DateTime', 'End DateTime',
                "Max Frequency", "Min Frequency",
                "Flare Location", "NOAA","Imp",
                "CME DateTime", "CPA", "Width",
                "Speed","Is Halo","Width Lower Bound"]] 
"""


dfNasa = dfNasa.drop(["Start Date", "Start Time", "End Date", "End Time", "CME Date", "CME Time"], axis = 1)
# And don't forget to check each column for the correct datatype
dfNasa.info()
dfNasa

```

### Question 1: Replication

The Nasa Data can replicate the SpaceWeather Table fairly well. It is missing a few values that have the same Imp values as other flare events. They are missing from the raw data on NASA's website, and it's hard to say whether the cause is a clerical error or some systemic error in the NASA data processing pipeline. Regardless, the result is that our replicated table includes some instances with Imp values that are lower than the min of the SpaceWeather Table.

```python
dfReplicate = dfNasa
dfReplicate["Imp Rank"] = [float(str(row[1:-1])) if "X" in str(row) else np.nan for row in dfReplicate["Imp"]]
dfRank = dfReplicate.sort_values(by="Imp Rank", ascending = False)
dfRank.head(50)
```

### Question 2: integration

```python
from datetime import timedelta as td
solarIter = dfSolar.iterrows() # sets up iterator for table
for i,row in solarIter:
    samePower = dfNasa[dfNasa['Imp'] == row['x_class']]
    if len(samePower) > 0:
        matchIter = samePower.iterrows()
        for j,r in matchIter:
            timeD1 = row['max_datetime'] - row['start_datetime'] 
            timeD2 = r['CME DateTime']- r['Start DateTime']
            if (timeD1 > timeD2+td(hours=-5)) and (timeD1 < timeD2+td(hours=5)):
                dfNasa.loc[j, 'Solar Rank'] = row['rank']
                break
dfNasa = dfNasa.sort_values(by='Solar Rank', ascending = True)
dfNasa.head(50)
```
