```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json 
import lxml
url = 'https://www.spaceweatherlive.com/en/solar-activity/top-50-solar-flares' # sets URL for easy use
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
# sets headers to mock an actual user^
response = requests.get(url, headers=headers) # sends get request to spaceweather and collects response
content = response.content # sets the content of the response
```

```python
a = BeautifulSoup(content) # create a beautifulsoup object for the content 
a.prettify() # prettify it 
solarTable = a.find('table') # finds the table on the website 
dfSolar = pd.read_html(solarTable.prettify()) # converts html table into pandas df
dfSolar=dfSolar[0] 
dfSolar.columns = ['rank','x_class', 'date', 'region' ,'start_time', 'max_time', 'end_time', 'movie'] # sets columns
dfSolar
```

### Step 2 - Tidy Flare Data

```python
del dfSolar['movie'] # cuts 'movie' column
```

```python
import datetime
iterr = dfSolar.iterrows() # sets up iterator for table
for i,row in iterr: # loops through rows
    
    temp = row['start_time'] + ' '+ row['date'] # create one string with both start time and date
    startDatetime = pd.to_datetime(temp) # convert string into a datetime object 
    dfSolar.loc[i,'start_time'] = startDatetime # uses current row index to replace time with datetime object 
    
    temp = row['max_time'] + ' '+ row['date'] # repeats replacement for all types of times
    maxDatetime = pd.to_datetime(temp)
    dfSolar.loc[i,'max_time'] = maxDatetime
    
    temp = row['end_time'] + ' '+ row['date']
    endDatetime = pd.to_datetime(temp)
    dfSolar.loc[i,'end_time'] = endDatetime

dfSolar = dfSolar.drop(columns=['date'], axis=1) # drops date column since no longer needed
dfSolar = dfSolar.rename(columns={"start_time": "start_datetime", "max_time": "max_datetime", "end_time": "end_datetime"}, errors="raise")
# renames columns ^
dfSolar = dfSolar[['rank','x_class','start_datetime','max_datetime','end_datetime','region']] # rearrange columns
dfSolar
```

### Step 3 - Scrape Nasa Data

```python
import re
import numpy as np
url2 = 'https://cdaw.gsfc.nasa.gov/CME_list/radio/waves_type2.html'
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
# sets headers to mock an actual user^
response2 = requests.get(url2, headers=headers) # sends get request to spaceweather and collects response
content2 = response2.content # sets the content of the response
```

```python
b = BeautifulSoup(content2, "html.parser") 
s = str(b.pre).split('\n')[12:-2] # removes top level html and skips to data 
data = {"Start Date": [], # sets up base series format to add to df
        "Start Time": [],
        "End Date": [],
        "End Time": [],
        "Start Frequency": [],
        "End Frequency": [],
        "Flare Location": [],
        "NOAA": [],
        "Imp": [],
        "CME Date": [],
        "CME Time": [],
        "CPA": [],
        "Width": [],
        "Speed": []}
        
dfNasa = pd.DataFrame(data) # creates empty df 
```

```python
for i in s:
    row = i.split(' ')
    a = [x for x in row if x] # finds non empty list elements
    b = [y for y in a if y!='<a'] # finds all elements that are not equal to a tag
    
    temp = re.sub(r'href[^<]+>', '', b[4]) # gets rid of url before number 
    noLinkMax = re.sub(r'</a>','', temp) # gets rid of a closing tag
    noLinkMax = re.sub(r'</a>','', temp) # in case of regex missing  
    if noLinkMax != '????': # checks if value is a num if so num is made into an int
        noLinkMax = int(noLinkMax)
    else:
        noLinkMax = np.nan # else made into a NaN
        
    
    temp = re.sub(r'href[^<]+>', '', b[5]) # same steps as LinkMax
    noLinkMin = re.sub(r'</a>','', temp)
    if noLinkMin != '????':
        noLinkMin = int(noLinkMin)
    else:
        noLinkMin = np.nan
    
    tempCME = re.sub(r'href[^<]+>', '', b[9]) # gets rid of url before CME Date 
    noLinkCME = re.sub(r'</a>','', tempCME) # removes closing a tag
    
    tempSpeed = re.sub(r'href[^<]+>', '', b[13]) # gets rid of url before speed 
    noLinkSpeed = re.sub(r'</a>','', tempSpeed)# removes closing a tag
    if noLinkSpeed != '----': # checks if value is a num if so num is made into an int
        noLinkSpeed = int(noLinkSpeed)
    else:
        noLinkSpeed = np.nan # else made into a NaN
    
    imp = b[8] # replaces . found at the end of some imp with + to match with top 50 data
    if imp[-1] == '.':
        imp = imp[:-1] + '+'
        
    year = b[0][0:4] # year of the current flare
    enddate = year + '/'+ b[2] # adds year to end and CME dates since those dates do not carry year
    
    cmeDate = year + '/' + noLinkCME
    
    insertRow = pd.Series({"Start Date": b[0], # sets all data to series columns and adds to df
                       "Start Time": b[1],
                       "End Date": enddate,
                       "End Time": b[3],
                       "Start Frequency": noLinkMax,
                       "End Frequency": noLinkMin,
                       "Flare Location": b[6],
                       "NOAA": b[7],
                       "Imp": imp,
                       "CME Date": cmeDate,
                       "CME Time": b[10],
                       "CPA": b[11],
                       "Width": b[12],
                       "Speed": noLinkSpeed
                      })
    dfNasa = dfNasa.append(insertRow, ignore_index=True)
dfNasa
```

### Step 4 - Tidy Nasa Data

```python
# This cell cleans NASA data by finding wrong/empty values and replacing them with right/NaN values
dfNasa["Flare Location"] = dfNasa["Flare Location"].replace("Back","BACK")
dfNasa["Flare Location"] = dfNasa["Flare Location"].replace("Back?","BACK")

dfNasa["NOAA"] = dfNasa["NOAA"].replace("-----",np.NaN)
dfNasa["NOAA"] = dfNasa["NOAA"].replace("DSF","FILA")
dfNasa["Imp"] = dfNasa["Imp"].replace("----",np.NaN)

dfNasa['CME Time'] = dfNasa['CME Time'].replace('--:--','00:00')

dfNasa["CME Date"] = dfNasa["CME Date"].replace("----",np.NaN)
dfNasa["CME Date"] = dfNasa["CME Date"].replace("----",np.NaN)
dfNasa["CPA"] = dfNasa["CPA"].replace("----",np.NaN)
dfNasa["Is Flare"] = [True if row == "Halo" else False for row in dfNasa['CPA']]
dfNasa["CPA"] = dfNasa["CPA"].replace("Halo",np.NaN)
dfNasa["Width"] = dfNasa["Width"].replace("----",np.NaN)
dfNasa["Width Lower Bound"] = [True if "&gt;" in str(row) else False for row in dfNasa['Width']]
dfNasa["Width"] = [int(str(row)[4:]) if "&gt;" in str(row) else row for row in dfNasa['Width']]

dfNasa["End Time"] = dfNasa["End Time"].replace("24:00","23:59")
dfNasa["Start DateTime"] = pd.to_datetime(dfNasa["Start Date"] + " " + dfNasa["Start Time"]) # creates datetime obj for each time related field
dfNasa["End DateTime"] = pd.to_datetime(dfNasa["End Date"] + " " + dfNasa["End Time"])
dfNasa["CME DateTime"] = pd.to_datetime(dfNasa["CME Date"] + " " + dfNasa["CME Time"])



dfNasa = dfNasa.drop(["Start Date", "Start Time", "End Date", "End Time", "CME Date", "CME Time"], axis = 1)
#drops unneeded columns
dfNasa.info() # displays dtypes of columns
dfNasa

```

### Question 1: Replication

The Nasa Data can replicate the SpaceWeather Table fairly well. It is missing a few values that have the same Imp values as other flare events. They are missing from the raw data on NASA's website, and it's hard to say whether the cause is a clerical error or some systemic error in the NASA data processing pipeline. Regardless, the result is that our replicated table includes some instances with Imp values that are lower than the min of the SpaceWeather Table.

```python
dfReplicate = dfNasa.copy() # copies Nasa df
dfReplicate["Imp Rank"] = [float(str(row[1:-1])) if "X" in str(row) else np.nan for row in dfReplicate["Imp"]]
# creates new field that X level flares and assigns them values based on their number
dfRank = dfReplicate.sort_values(by="Imp Rank", ascending = False) # sorts largest to smallest on new rank
dfRank.head(50) # top 50 flares by just their power rank
```

### Question 2: integration

We decided our matching algorithm would be layered and based on two factors: the x_class/imp of a flare and the timedelta between the start time and the max/CME time. First we wanted to match the imp/x_class becuase the power of the flare seems to be the most distingushing factor of a flare on the Top 50 list. Second, we take the differences between the start datetime and max/cme datetime of each potential match and compare them. If the time deltas are within 5 hours of each other, then its a match! We chose 5 hours because there seems to be major differences between the two sources of data but they seem to be with in 5 hours from the ones we know for sure are a match. 

```python
from datetime import timedelta as td
solarIter = dfSolar.iterrows() # sets up iterator for table
for i,row in solarIter: 
    samePower = dfNasa[dfNasa['Imp'] == row['x_class']] # finds all flares from Nasa with same power level
    if len(samePower) > 0:
        matchIter = samePower.iterrows()
        for j,r in matchIter:
            timeD1 = row['max_datetime'] - row['start_datetime']  # finds differences in CME and end time
            timeD2 = r['CME DateTime']- r['Start DateTime']
            if (timeD1 > timeD2+td(hours=-5)) and (timeD1 < timeD2+td(hours=5)): 
                # compares differences in time deltas to see if the flare close to the SolarWeather flares
                dfNasa.loc[j, 'Solar Rank'] = row['rank'] 
                break
dfNasa = dfNasa.sort_values(by='Solar Rank', ascending = True) # assign new ranking and sort 
dfNasa.head(25) # show top 50 -> we only see 25 matches
```

### Question 3: Analysis Plot

#### Ploting Speed Over Time

For our analysis plot we wanted to see if there was any covariation between the speed of the top 50. We wanted to see if the top 50 flares were skewed towards any one time interval

```python
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.dates as mdates

# Plot of Speed with respect to CME Date Time
dfMerge = dfNasa.head(25) # only ranked flares

dfReIndex = dfMerge.set_index("CME DateTime") # changes index to datetime of CME
dfTop50 = dfReIndex.sort_index() # sorts the index 

dfNasaReIndex = dfNasa.set_index("CME DateTime") # does same change index and sorting
dfNasaSort = dfNasaReIndex.sort_index()
```

```python
y = dfNasaSort["Speed"] # plot speed over index (time)
x = dfNasaSort.index

ypt = dfTop50["Speed"] # mark the top 50 flares 
xpt = dfTop50.index

plt.xlabel('Time') # labels of plot
plt.ylabel('Speed')

plt.plot(x,y) # plot both Xs and Ys
plt.plot(xpt, ypt, "rx")
```

This is a plot of speed over CME datetimes across the Nasa data set

From our plot we can see that there seems to be slight skew towards before 2005 of top 50 flares across time. We do see a gap between 2009 and 2011 which may be a cause for the skew. Additionally, from our plot we can see that the top 50 flares seem to be faster on average but not always the fastest. This definitly points towards there being a positive covarience between speed and strength.
